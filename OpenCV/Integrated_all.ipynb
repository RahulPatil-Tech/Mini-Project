**Import Libraries**

```python
from tkinter import *
from PIL import Image, ImageTk
import cv2
import numpy as np
import pandas as pd
import xlsxwriter
```

**This will load video** 

```python
cap=cv2.VideoCapture('1.mp4')
set_width = 640
set_height = 480
```

**Set Height Width of Cam**

```python
width, height = 640, 480
cap.set(cv2.CAP_PROP_FRAME_WIDTH, width)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, height)
```

**GUI app with escape function to close app**

```python
app=Tk()
app.bind=('<Escape>', lambda e: app.quit())
```

**Create label where live video will get displayed**

```python
label_width=Label(app)
label_width.pack()
```

***Angle Detection Function***

```python
while True:
    # Read the frame from the video
    ret, frame = cap.read()

    # Convert the frame to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Apply Canny edge detection to find edges
    edges = cv2.Canny(gray, 50, 150, apertureSize=3)

    # Apply Hough Line Transform to detect lines
    lines = cv2.HoughLines(edges, 1, np.pi/180, 200)

    # Loop through each line and calculate the angle
    if lines is not None:
        for line in lines:
            rho, theta = line[0]
            angle = theta * 180/np.pi
            print('Angle:', angle)

            # Draw the line on the original frame
            a = np.cos(theta)
            b = np.sin(theta)
            x0 = a * rho
            y0 = b * rho
            x1 = int(x0 + 100*(-b))
            y1 = int(y0 + 100*(a))
            x2 = int(x0 - 100*(-b))
            y2 = int(y0 - 100*(a))
            cv2.line(frame, (x1,y1), (x2,y2), (0,0,255), 2)
            df = pd.DataFrame({'Angle': [angle]})
            with open('data.csv', 'a') as f:
                df.to_csv(f, header=False, index=False)
    # Display the frame with detected lines
    cv2.imshow('Lines', frame)
    cv2.imshow('Edges', edges)

    # Exit the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```

**Function to open camera**

```python
def open_camera():
    _,frame=cap.read()
    frame=cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    frame=Image.fromarray(frame)
    frame=ImageTk.PhotoImage(frame)
    label_width.config(image=frame)
    label_width.image=frame
    label_width.after(10, open_camera)
```

**Button for open cam**

```python
button1=Button(app, text='Open Camera', command=open_camera)
button1.pack()
app.mainloop()
```
